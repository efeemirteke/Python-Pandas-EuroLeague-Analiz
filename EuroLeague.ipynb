```python
import pandas as pd
```

```python
db=pd.read_csv("euroleaugeTeams.csv")
```

```python
db.drop(db[db["season_code"]!="E2024"].index,inplace=True)
db.head(1)
```

```python
def yeniTakımIsımleri(row):
    if row["team_id"] == "ASV":
        return "LDLC ASVEL Villeurbanne"
    elif row["team_id"] == "BAR":
        return "FC Barcelona"
    elif row["team_id"] == "BAS":
        return "Cazoo Baskonia Vitoria-Gasteiz"
    elif row["team_id"] == "BER":
        return "ALBA Berlin"
    elif row["team_id"] == "IST":
        return "Anadolu Efes İstanbul"
    elif row["team_id"] == "MAD":
        return "Real Madrid"
    elif row["team_id"] == "MCO":
        return "AS Monaco"
    elif row["team_id"] == "MIL":
        return "Olimpia Milano"
    elif row["team_id"] == "MUN":
        return "FC Bayern Munich"
    elif row["team_id"] == "OLY":
        return "Olympiacos Piraeus"
    elif row["team_id"] == "PAN":
        return "Panathinaikos Athens"
    elif row["team_id"] == "PAR":
        return "Paris Basketball"
    elif row["team_id"] == "PRS":
        return "Partizan Mozzart Bet Belgrade"
    elif row["team_id"] == "RED":
        return "Crvena Zvezda Meridianbet Belgrade"
    elif row["team_id"] == "TEL":
        return "Maccabi Tel Aviv"
    elif row["team_id"] == "ULK":
        return "Fenerbahçe Beko İstanbul"
    elif row["team_id"] == "VIR":
        return "Virtus Segafredo Bologna"
    elif row["team_id"] == "ZAL":
        return "Zalgiris Kaunas"
    else:
        return "Bilinmiyor"

db["takim_isimleri"] = db.apply(yeniTakımIsımleri, axis=1)
```
